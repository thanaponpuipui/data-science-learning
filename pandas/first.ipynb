## manipulate data with pandas

```python
# create new datas to work with
food_missing = {
    'Name': ['curry', 'beef blow', 'frame grill burger', 'kapao rice', 'mix vegetable fried rice', 'tomato soup', 'onion fried'],
    'Size per serve': ['250g', np.nan, '200g', '170g', '200g', np.nan, '100g'],
    'Price': ['$10.00', '$13.00', '$5.50', np.nan, '$8.00', np.nan, np.nan],
    'Serve per day': [30, 25, 55, 23, 3, 0, 80]
}
foods = pd.DataFrame(data=food_missing)
```

```python
foods
```

```python
foods.dropna()
```

```python
foods
```

```python
size_int = foods['Size per serve'].str.replace('g', '').dropna().astype(int)
size_int
```

```python
size_int.mean()
```

```python
# fill NaN in size by mean
foods['Size per serve'] = foods['Size per serve'].fillna(size_int.mean().astype(str).replace('.0', 'g'))
foods
```

```python
# fill all NaN in Price Individually
foods.loc[foods['Name'] == 'kapao rice', 'Price'] = '$11.00'
foods.loc[foods['Name'] == 'onion fried', 'Price'] = '$3.00'
foods.loc[foods['Price'].isnull(), 'Price'] = '$5.50'
foods
```
